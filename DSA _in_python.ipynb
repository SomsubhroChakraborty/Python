# Linked List

```python
linked_list = []  # Create an empty linked list

# Add some string type elements to the linked list
linked_list.append("America")
linked_list.append("India")
linked_list.append("Japan")

print(linked_list)

# Display menu
choice = 0
while choice < 5:
    print("Operation you want to do")
    print("1 to add element in Linkedlist")
    print("2 to remove element from Linkedlist")
    print("3 to replace element in Linkedlist")
    print("4 to Search element in Linkedlist")
    print("5 to exit")
    choice = int(input("Enter your choice"))

    # Operations
    if choice == 1:
        element = input("Enter the element you want to insert : ")
        pos = int(input("Enter the position"))
        linked_list.insert(pos, element)
    elif choice == 2:
        element = input("Enter the element you want to remove : ")
        linked_list.remove(element)
    elif choice == 3:
        element = input("Enter the element you want to replace with : ")
        pos = int(input("Which position you want to replace"))
        linked_list[pos] = element
    elif choice == 4:
        element = input("Enter the element you want to search")
        if element in linked_list:
            pos = linked_list.index(element)
            print("Element found at position:", pos)
        else:
            print("Element not found in linked list")
    else:
        break
    print(linked_list)
```

# Stack

```python
class Stack:
    def __init__(self):
        self.items = []
    def push(self, element):
        self.items.append(element)
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1
    def peep(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None
    def search(self, element):
        if element in self.items:
            return self.items.index(element)
        else:
            return -1
    def is_empty(self):
        return len(self.items) == 0

s = Stack()  # create empty stack

# display menu
choice = 0
while choice < 5:
    print("Choose the stack operation you want to perform")
    print("1 for Push")
    print("2 for Pop")
    print("3 for Peep")
    print("4 for Search")
    print("5 for Exit")
    choice = int(input("Enter your choice : "))

    if choice == 1:
        element = input("Enter the element : ")
        s.push(element)
    elif choice == 2:
        element = s.pop()
        if element == -1:
            print("Stack is empty")
        else:
            print("Popped element is:", element)
    elif choice == 3:
        element = s.peep()
        if element is not None:
            print("Top element is:", element)
        else:
            print("Stack is empty")
    elif choice == 4:
        element = input("Enter the element : ")
        pos = s.search(element)
        if pos != -1:
            print("Element found at position:", pos)
        else:
            print("Element not found in stack")
    else:
        break
```

# Queue

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, element):
        self.items.append(element)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return -1

    def front(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def search(self, element):
        if element in self.items:
            return self.items.index(element)
        else:
            return -1

    def is_empty(self):
        return len(self.items) == 0

q = Queue()  # create empty queue

# display menu
choice = 0
while choice < 6:
    print("Choose the queue operation you want to perform")
    print("1 for Enqueue")
    print("2 for Dequeue")
    print("3 for Front")
    print("4 for Rear")
    print("5 for Search")
    print("6 for Exit")
    choice = int(input("Enter your choice : "))

    if choice == 1:
        element = input("Enter the element : ")
        q.enqueue(element)
    elif choice == 2:
        element = q.dequeue()
        if element == -1:
            print("Queue is empty")
        else:
            print("Dequeued element is:", element)
    elif choice == 3:
        element = q.front()
        if element is not None:
            print("Front element is:", element)
        else:
            print("Queue is empty")
    elif choice == 4:
        element = q.rear()
        if element is not None:
            print("Rear element is:", element)
        else:
            print("Queue is empty")
    elif choice == 5:
        element = input("Enter the element : ")
        pos = q.search(element)
        if pos != -1:
            print("Element found at position:", pos)
        else:
            print("Element not found in queue")
    else:
        break
```

# Quicksort

```python
def quicksort(arr,lo,hi):
  if(lo < hi):
    p = partition(arr,lo,hi)
    quicksort(arr,lo,p-1)
    quicksort(arr,p+1,hi)
def partition(arr,lo,hi):
  pivot = arr[hi]
  i = lo - 1
  for j in range(lo,hi):
    if(arr[j] <= pivot):
      i +=1
      arr[i],arr[j]=arr[j],arr[i]
  arr[i+1], arr[hi] = arr[hi],arr[i+1]
  return i+1

arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quicksort(arr, 0, n-1)
print("Sorted array:", arr)
```

#Mergesort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Example usage:
arr = [12, 11, 13, 5, 6, 7,4,2,78,3,38,23,1,0]
merge_sort(arr)
print("Sorted array:", arr)
```

# Binary search

```python
items = [2, 3, 4, 10, 40]
target = int(input("Enter the target number :"))
low = 0
high = len(items) - 1
found = False
while low <= high and not found:
    mid = (low + high) // 2
    if items[mid] == target:
        found = True
        print("Item found at index:", mid)
    elif items[mid] < target:
        low = mid + 1
    else:
        high = mid - 1

```
