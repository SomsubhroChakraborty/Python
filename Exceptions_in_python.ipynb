## **Explain Python Built-in-Exceptions.**

### In Python, built-in exceptions are predefined classes that represent various types of errors or exceptional conditions that can occur during program execution. These exceptions are raised (or thrown) when an error occurs in Python code, and they can be caught (or handled) by using exception handling mechanisms like try and except blocks.

**SyntaxError:** Raised when the Python parser encounters a syntax error in the code. It indicates that the code is not syntactically valid Python.

**IndentationError:** Raised when there's an improper indentation in the code, typically due to inconsistent use of spaces or tabs.

NameError: Raised when an identifier (variable name, function name, etc.) is not found in the local or global scope.

TypeError: Raised when an operation or function is applied to an object of an inappropriate type.

ValueError: Raised when a built-in operation or function receives an argument with the correct type but an inappropriate value.

KeyError: Raised when a dictionary key is not found in the set of existing keys.

IndexError: Raised when a sequence subscript (index) is out of range.

FileNotFoundError: Raised when a file or directory is requested but cannot be found.

ZeroDivisionError: Raised when division or modulo operation is performed with zero as the divisor.

AttributeError: Raised when an attribute reference or assignment fails because the specified attribute doesn't exist.

ImportError: Raised when an import statement fails to find the module, or when a from ... import ... statement cannot find the specified name in the module.

KeyboardInterrupt: Raised when the user interrupts the execution of the program (e.g., by pressing Ctrl+C).

## Differentiate between Syntax error and exception

Syntax errors occur during the parsing (or compilation) phase of Python code, before the code is executed.
They are caused by violations of the Python syntax rules, such as incorrect indentation, missing or misplaced punctuation, or incorrect keyword usage

Exceptions, on the other hand, occur during the execution phase of Python code, after the code has been successfully parsed and is being executed.
Exceptions represent runtime errors or exceptional conditions that occur while the program is running, such as division by zero, attempting to access an index that is out of range, or trying to open a file that does not exist.
Exceptions can be caused by a variety of factors, including user input, external resources (e.g., files, network connections), and program logic.

## How exception handling works in Python?


Exception handling in Python allows you to handle errors that occur during the execution of a program gracefully, preventing the program from crashing abruptly. Here's how it works:

**Try block: **You place the code that might raise an exception inside a try block. If an exception occurs within this block, Python searches for an appropriate except block to handle it.

**Except block:** If an exception occurs within the try block, Python looks for an except block that matches the type of exception raised. If a matching except block is found, its code is executed.
**Finally block (optional):** The finally block is always executed, regardless of whether an exception occurred in the try block. This block is useful for cleaning up resources, such as closing files or database connections.

```python
try:
    # code that might raise an exception
    # e.g., division by zero or accessing a non-existent index
except SomeException:
    # handle SomeException
except AnotherException:
    # handle AnotherException
else:
    # optional block that executes if no exception occurs
finally:
    # optional block that always executes, whether an exception occurred or not
```
