# **What is Inheritance in python**

Inheritance is the property of one class to use the properties of its parrent class.

**BENIFITS**
1. represents real world relationship.
2.provides reusability.

````python
#single inheritance
class P:
  def m1(self):
    print("Parent class")
class c(P):
  def m2(self):
    print("Child class")
C=c()
C.m1()
C.m2()
````

````python
#multilevel inheritance
class P:
  def m1(P):
    print("Parent class")
class C(P):
  def m2(self):
    print("Child class")
class GC(C):
  def m3(self):
    print("grand child")
gc = GC()
gc.m1()
gc.m2()
gc.m3()
````

````python
#multiple inheritance
class P1:
  def m1(P1):
    print("parent class 1")
class P2:
  def m2(P2):
    print("parent class 2")
class C(P1, P2):
  def m3(self):
    print("child class")
c = C()
c.m1()
c.m2()
c.m3()
````

# **What is Polymorphism**

## polymorphism is the proces where we can use a single function with different functionality and same name

````python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Polymorphism in action
animals = [Dog(), Cat()]
for animal in animals:
    animal.sound()
````

# **Encapsulation**

## Encapsulation describes the wrapping of data and methods within one unit .This put restriction of accessing data and prevent them for accidental changes.

**Public:** By default, all attributes and methods in Python are public, meaning they can be accessed from outside the class.

**Protected:** Conventionally, attributes and methods prefixed with a single underscore '_' are considered protected, indicating that they should not be accessed from outside the class, but this is just a convention and not enforced by the language.

**Private:** Similarly, attributes and methods prefixed with a double underscore '__' are considered private, indicating that they should not be accessed or overridden directly from outside the class. However, Python uses name mangling to make it difficult but not impossible to access these attributes and methods from outside the class.

````python
class MyClass:
    def __init__(self):
        self.public_attr = 10
        self._protected_attr = 20
        self.__private_attr = 30

    def public_method(self):
        print("Public method")

    def _protected_method(self):
        print("Protected method")

    def __private_method(self):
        print("Private method")

# Accessing attributes and methods from outside the class
obj = MyClass()
print(obj.public_attr)
print(obj._protected_attr)
# The following line would raise an AttributeError because __private_attr is name-mangled
# print(obj.__private_attr)
obj.public_method()
obj._protected_method()
# The following line would raise an AttributeError because __private_method is name-mangled
# obj.__private_method()
````

Here we use `"__init__"` whisch is used as constructor in python
