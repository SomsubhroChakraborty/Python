## Create a string with a value ‘Computer Science’. Print the string as “cOMPUTER sCIENCE”

```python
original_string = 'Computer Science'

swapped_case_string = ''
for char in original_string:
  if char.isupper():
        swapped_case_string += char.lower()
  elif char.islower():
        swapped_case_string += char.upper()
  else:
        swapped_case_string += char

print('"' + swapped_case_string + '"')
```
